```python
import numpy as np
from spatialmath import *
from math import pi

import matplotlib.pyplot as plt
%matplotlib notebook 
```

```python
H0 = SE3.Tx(1.0) * SE3.Ty(1.0) * SE3.Tz(0)
H1 = SE3.Tx(3.0) * SE3.Ty(1.0) * SE3.Tz(0) * SE3.Rz(-45, 'deg')
H2 = SE3.Tx(3.0) * SE3.Ty(1.0) * SE3.Tz(2) * SE3.Rz(-45, 'deg')* SE3.Ry(45, 'deg')
H3 = SE3.Tx(1.0) * SE3.Ty(1.0) * SE3.Tz(2) * SE3.Rz(-45, 'deg')* SE3.Ry(45, 'deg')* SE3.Rx(45, 'deg')
Hfinal = SE3.Tx(1.0) * SE3.Ty(1.0) * SE3.Tz(4) * SE3.RPY(45,45,-45, unit='deg')
```

```python
%matplotlib inline
H0.plot(frame='0', color='black', dims=[0,4])
H1.plot(frame='1', color='blue', dims=[0,4])
H2.plot(frame='2', color='green', dims=[0,4])
H3.plot(frame='3', color='navy', dims=[0,4])
Hfinal.plot(frame='final', color='red', dims=[0,4])

#The Translation here is to better show the changes in rotation. We want to see how rotation changes the object orientation for Hfinal. 
#We can see that it is first rotated about the Z axis then Y axis then X axis. Here the order of the rotation matters as the new rotation 
#is calculated from the current axis orientation.
```
